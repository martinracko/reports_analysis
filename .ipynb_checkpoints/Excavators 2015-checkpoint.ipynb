# Crawler Excavators 2015 Pre-analysis

Duplicates excluded, listings + auctions with and w/o prices from Europe. Based on ~43K records, whole world counts around 71K records.

```python
import json
records = [json.loads(line) for line in open('excavators.json')]
from pandas import DataFrame, Series
import pandas as pd; import numpy as np
import matplotlib as plt
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['savefig.dpi'] = 100
```

```python
frame = DataFrame(records)
frame.c = frame.c.astype(float64)
```

## Portal Breakdown

```python
by_p = frame.groupby(['portal', frame['auction']])
agg = (by_p.size().unstack().fillna(0)).sort(axis=0)
agg.plot(kind='barh', stacked=True)
```

## All Crawler Excavators

```python
by_q = frame.groupby(['q', frame['auction']])
agg_counts = (by_q.size().unstack().fillna(0)).sort(axis=1)
agg_counts.plot(kind='bar', stacked=True)
```

## Range (5.7, 12]t

```python
frame_ex = (frame[frame.c > 5.7])[frame.c <= 12]
by_q = frame_ex.groupby(['q', frame_ex['auction']])
agg_counts = (by_q.size().unstack().fillna(0)).sort(axis=1)
agg_counts.plot(kind='bar', stacked=True)
```
