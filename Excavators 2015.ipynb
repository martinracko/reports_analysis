# Crawler Excavators 2015 Pre-analysis

Duplicates excluded, listings + auctions mapped to model reliably from Europe. Based on ~43K records, whole world counts around 71K records.

```python
import json
records = [json.loads(line) for line in open('excavators.json')]
from pandas import DataFrame, Series
import pandas as pd; import numpy as np
import matplotlib as plt
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['savefig.dpi'] = 100
```

```python
frame = DataFrame(records)
frame.c = frame.c.astype(float64)
idx = pd.Index([u'2010Q1', u'2010Q2', u'2010Q3', u'2010Q4', \
                u'2011Q1', u'2011Q2', u'2011Q3', u'2011Q4', \
                u'2012Q1', u'2012Q2', u'2012Q3', u'2012Q4', \
                u'2013Q1', u'2013Q2', u'2013Q3', u'2013Q4', \
                u'2014Q1', u'2014Q2', u'2014Q3'])
```

## Portal Breakdown

```python
by_p = frame.groupby(['portal', frame['auction']])
agg_counts = (by_p.size().unstack().fillna(0))
agg_counts.plot(kind='barh', stacked=True)
```
