# Tractors 2015 Pre-analysis

Duplicates included for portals overview, excluded for category class overviews, listings + auctions mapped to model reliably from Europe. Based on ~180K records (~108K w/o duplicates). Power converted from kW to HP (1 kilowatt = 1.34102209 HP)

```python
import json
records = [json.loads(line) for line in open('tractors_dup.json')]
from pandas import DataFrame, Series
import pandas as pd; import numpy as np
import matplotlib as plt
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['savefig.dpi'] = 100
```

```python
frame = DataFrame(records)
frame.c = frame.c.astype(float64)
idx = pd.Index([u'2010Q1', u'2010Q2', u'2010Q3', u'2010Q4', \
                u'2011Q1', u'2011Q2', u'2011Q3', u'2011Q4', \
                u'2012Q1', u'2012Q2', u'2012Q3', u'2012Q4', \
                u'2013Q1', u'2013Q2', u'2013Q3', u'2013Q4', \
                u'2014Q1', u'2014Q2', u'2014Q3'])
```

## Portal Breakdown

```python
frame.groupby(['portal', frame['auction']]).size().unstack().fillna(0).plot(kind='barh', stacked=True)
```

## All Tractors

```python
records = [json.loads(line) for line in open('tractors.json')]
frame = DataFrame(records)
frame.c = frame.c.astype(float64)
frame.groupby(['q', frame['auction']]).size().unstack().fillna(0).plot(kind='bar', stacked=True)
```

## Range (0, 50]HP

```python
frame_ex = (frame[frame.c > 0])[frame.c <= 50]
frame_ex.groupby(['q', frame_ex['auction']]).size().unstack().fillna(0).sort(axis=1).plot(kind='bar', stacked=True)
```

## Range (50, 100]HP

```python
frame_ex = (frame[frame.c > 50])[frame.c <= 100]
frame_ex.groupby(['q', frame_ex['auction']]).size().unstack().fillna(0).plot(kind='bar', stacked=True)
```

## Range (100, 200]HP

```python
frame_ex = (frame[frame.c > 100])[frame.c <= 200]
frame_ex.groupby(['q', frame_ex['auction']]).size().unstack().fillna(0).plot(kind='bar', stacked=True)
```

## Range (200, -)HP

```python
frame_ex = (frame[frame.c < 200])
frame_ex.groupby(['q', frame_ex['auction']]).size().unstack().fillna(0).plot(kind='bar', stacked=True)
```
